```python
#customised fun is used to sort the images in numeric order
import re

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [ atoi(c) for c in re.split('(\d+)', text) ]
```

```python
#reading both train and test data
img_data = pd.read_csv("Images_Train.csv")
img_test = pd.read_csv("Images_Test.csv")
```

```python
# Test data had 2 categories "Sports and Team Jersey" and "Sports". Since they are same, clubbing them to sports
img_test.replace('Sports and Team Jersey','Sports',inplace=True)
```

```python
#converting Sub_category attribute as category data type
img_data['Sub_category'] = img_data['Sub_category'].astype('category')
img_test['Sub_category'] = img_test['Sub_category'].astype('category')
y_train = img_data['Sub_category']
y_test_data = img_test['Sub_category']
```

```python

y_test_data.shape
```

```python
#converting labels to numeric 
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y_train = le.fit_transform(y_train)
y_test_data = le.fit_transform(y_test_data)
```

```python
np.unique(y_test_data)
```

```python
from keras.utils import to_categorical
```

```python
y_true_tst = y_test_data
y_train = to_categorical(y_train)
y_test_data = to_categorical(y_test_data)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
X_train[0].shape
```

```python
num_classes = 22
input_shape = X_train[0].shape
num_epoch = 10
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
```

```python
#building model 
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape,padding='same'))
model.add(Conv2D(32, (3, 3), activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))

model.add(Conv2D(64, (3, 3), activation='relu',padding='same'))
model.add(Conv2D(64, (3, 3), activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))

model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adadelta', metrics=["accuracy"])
```

```python
model.summary()
```

```python
#built model using adadelta as optimizer. got acc of 55.98 for train and 55.94 for val
hist = model.fit(X_train, y_train, batch_size=32, epochs=num_epoch, verbose=1, validation_data=(X_test, y_test))
```

```python
#building same model using rmsprop as optimizer
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=["accuracy"])
```

```python
#got acc of 72.97 for train and 62.48 for val. Number of epochs used to train was 2
hist_1 = model.fit(X_train, y_train, batch_size=32, epochs=2, verbose=1, validation_data=(X_test, y_test))
```

```python
# Custom_vgg_model_1
#Training the classifier alone
from keras.layers import Input, Dense
image_input = Input(shape=(224, 224, 3))
```

```python
from keras.applications.vgg16 import VGG16

model = VGG16(input_tensor=image_input, include_top=True, weights='imagenet')
```

```python
model.summary()
```

```python
last_layer = model.get_layer('fc2').output
out = Dense(num_classes, activation='softmax', name='output')(last_layer)
```

```python
from keras.models import Model

custom_vgg_model = Model(image_input, out)
custom_vgg_model.summary()
```

```python
for layer in custom_vgg_model.layers[:-1]:
    layer.trainable = False
```

```python
custom_vgg_model.compile(loss='categorical_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```python
m3 = custom_vgg_model.fit(X_train, y_train, batch_size=16, epochs=2, verbose=1, validation_data=(X_test, y_test))
```
